```python
import pandas as pd
import plotly.express as px

#please use this cell to read and select your data
df = pd.read_csv('/content/drive/MyDrive/olympic_medals.csv')

# Filter the data for years since 2000
df = df[df['Year'] >= 2000]

```

```python
#Please use this cell to create your your figure. Please use Year column to color your graph.
import plotly.express as px
df = df[df['Year'] >= 2000]

fig = px.parallel_categories(df, 
                             dimensions=['Gender', 'Medal'], 
                             color='Year',
                             color_continuous_scale=px.colors.sequential.Inferno,
                             title='Proportions of Medal Type for Each Gender since year 2000')

fig.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
 
# create DataFrame
df = pd.DataFrame({'High Temp': [28, 30, 34, 38, 45, 42,
                                 38, 35, 32, 28, 25, 21],
                   'Low Temp': [22, 26, 30, 32, 41, 38,
                                32, 31, 28, 22, 15, 15],
                   'Avg Temp': [25, 28, 32, 35, 43, 40,
                                35, 33, 30, 25, 20, 18]},
                  index=['Jan', 'Feb', 'Mar', 'Apr', 'May',
                         'Jun', 'Jul', 'Aug', 'Sep', 'Oct',
                         'Nov', 'Dec'])
 
 
# create stacked bar chart for monthly temperatures
df.plot(kind='bar', stacked=True, color=['red', 'skyblue', 'green'])
 
# labels for x & y axis
plt.xlabel('Months')
plt.ylabel('Temp ranges in Degree Celsius')
 
# title of plot
plt.title('Monthly Temperatures in a year')
```

```python
#You can use this cell to write your code. It is doable at most 4 lines of code. 
import pandas as pd
import matplotlib.pyplot as plt

# read the CSV file
df = pd.read_csv('/content/drive/MyDrive/2016elections.csv', usecols=['state', 'st', 'total_vote', 'pct_clinton', 'pct_trump', 'pct_johnson', 'pct_other'])

# select rows for AR, MI, CA, and WI
df = df[df['st'].isin(['AR', 'MI', 'CA', 'WI'])]

# create stacked bar chart for vote percentages
df.plot(kind='bar', x='st', y=['pct_clinton', 'pct_trump', 'pct_johnson', 'pct_other'], stacked=True, color=['blue', 'red', 'orange', 'green'])

# labels for x & y axis
plt.xlabel('States')
plt.ylabel('Vote Percentages')
 
# title of plot
plt.title('2016 US Presidential Election Results')

# show the plot
plt.show()


```
