```python

titanic = pd.read_csv('/content/gdrive/MyDrive/titanic.csv')
# Rename 1s in the 'Survived' column to 'Survived'
titanic['Survived'] = titanic['Survived'].replace({1: 'Survived'})

# Get rows that do not have "?" under the 'Pclass' column
titanic_filtered = titanic[titanic['Pclass'] != '?']


```

Create a multipanel plot which draws density plots for each class-survived combination. Please also use the hue parameter based on the Sex attribute to have two density plots in each panel with different colors.

```python
# Change 1s to "Survived" under the Survived column
titanic.loc[titanic['Survived'] == 1, 'Survived'] = "Survived"

# Filter rows without "?" under Pclass column
titanic = titanic[titanic['Pclass'] != "?"]
g = sns.FacetGrid(titanic, col="Survived", row="Pclass", hue="Sex", height=3, aspect=1.2)
g.map(sns.kdeplot, "Age", linewidth=1.5)
g.add_legend()
```
